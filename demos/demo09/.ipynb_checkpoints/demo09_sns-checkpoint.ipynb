# 2. One Qualitative Variable

When we have qualitative categorical variables, we are usually interested in their counts. We want to see how many data observations are present in each category.

```python
#### Adding a random 'city' column to our generated data ####

cities = ['Tokyo', 'Porto', 'New York', 'Paris', 'Bern']
rand_cities = np.random.choice(cities, 167)
generated_data['city'] = rand_cities
generated_data.head()
```

```python
#### Adding another qualitative variable ####

smokes = ['smoking', 'not smoking', 'social smoker']
rand_s = np.random.choice(smokes, 167, p = [0.5, 0.3, 0.2])
generated_data['smoking status'] = rand_s
generated_data.head()
```

```python
counts_sm = generated_data['smoking status'].value_counts().to_frame().reset_index()
counts_sm
```

### Pie Charts

```python
plt.pie(counts_sm['smoking status'], labels=counts_sm['index'], autopct='%1.1f%%')
```

### Barplots 

```python
sns.countplot(x = 'smoking status', data = generated_data)
```

```python
sns.countplot( x= 'city', data = generated_data)
```

### Word Clouds

* For more textual visualizations visit http://textvis.lnu.se/

```python
#### visualizing through words ####
city_string = ' '.join(list(generated_data['city'])) 
wc = WordCloud().generate(city_string)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
```

```python
#### creating counts df for the cities ####
counts_city = generated_data['city'].value_counts().to_frame().reset_index().rename(columns = {'index': 'city name', 'city' : 'counts'})
counts_city
#### barplot ####
f = sns.barplot(x="city name", y="counts", data = counts_city)
f.set_xticklabels(f.get_xticklabels(), rotation=30)
```

### Stripplots

```python
f = sns.stripplot(x="city name", y="counts", data = counts_city)
f.set_xticklabels(f.get_xticklabels(), rotation=30)
```

# 3. One Quantitative Variable and One Qualitative Variable

* Comparing quantitative information across the categories.

```python
sns.catplot(x='city', y='life expectancy', data=generated_data)
```

```python
for i in set(['New York', 'Porto']):
    sns.distplot(generated_data.loc[generated_data['city'] == i, 'life expectancy'], hist=True, kde=False, label=i)
plt.legend()
plt.title("Overlaid Life Expectancy Distributions of New York and Porto")
```

```python
for i in set(['New York', 'Porto']):
    sns.distplot(generated_data.loc[generated_data['city'] == i, 'life expectancy'], hist = False, label=i)
plt.legend()
plt.title("Overlaid Life Expectancy Distributions of New York and Porto")
```

### Side-to-side Boxplots

```python
f = sns.boxplot(x='city', y='life expectancy', notch=False, data=generated_data)
f.set_xticklabels(f.get_xticklabels(), rotation=90)
```

### Side-to-side Violin plots

```python
f = sns.violinplot(x="city", y="life expectancy",data=generated_data)
f.set_xticklabels(f.get_xticklabels(), rotation=30)
```
